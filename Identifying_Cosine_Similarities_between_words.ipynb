####Cosine Similarities and TF-IDF Calculation

```python
#importing the necessary packages to run the code
import pandas as pd 
import numpy as np 
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#keeping the file names in a single list to access them easily later
file_name=['01.txt','02.txt','03.txt','04.txt','05.txt','06.txt','07.txt','08.txt','09.txt','10.txt','11.txt','12.txt','13.txt','14.txt',
          '15.txt','16.txt','17.txt','18.txt','19.txt','20.txt']
```

```python
#data processing and converting into legible data
eachword_frequency = []

#reading the file and pre processing the data
def data_processing(text_file):
    actual_text = text_file.read()
    lowercase = actual_text.lower()
    punctuated=remove_punctuation(lowercase)
    words = punctuated.split()
    final_words = [word.replace("'s", '') for word in words]
    return final_words

#removing punctuation marks and unnecessary symbols
def remove_punctuation(sym):
    punctuation_marks='!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'#special symbols 
    for w in punctuation_marks:
        sym = sym.replace(w,'')
    return sym


#identifying unique words in the data
def unique_words():
    unique_words = []
    for i in range(len(file_name)):
        text_file = open('C:/Users/akhil/Downloads/Assignment1_data/Assignment1_data/'+file_name[i], 'r')
        words = data_processing(text_file)
        for word in words:
            if word not in unique_words:
                unique_words.append(word)
    return unique_words

#calculating the frequency of the unique words
def frequency_words(unique_words):
    words_frequency = {}
    for i in unique_words:
        word_count = []
        for j in range(len(file_name)):
            text_file = open('C:/Users/akhil/Downloads/Assignment1_data/Assignment1_data/'+file_name[j], 'r')
            words = data_processing(text_file)
            word_count.append(words.count(i))
        key,value = i,word_count
        words_frequency[key] = word_count
        eachword_frequency.append(word_count)
    return words_frequency,eachword_frequency
```

```python
unique = unique_words()
frequency,eachword_frequency = frequency_words(unique)
```

```python
#3 calculating the TF IDF values
#converting the individual files into a consolidated file and using the tfid vectorizer to fing out the tfidf values
consolidated_text = []
for i in range(20):
    text_file = open('C:/Users/akhil/Downloads/Assignment1_data/Assignment1_data/'+file_name[i], 'r')
    text = text_file.read()
    consolidated_text.append(text)
def vectorized(consolidated_text):
    vectorizer = TfidfVectorizer()
    tfid = vectorizer.fit_transform(consolidated_text)
    result = dict(zip(vectorizer.get_feature_names(), tfid.toarray()[0]))
    return result
vectorized(consolidated_text)
```

```python
#4 finding the cosine similarity between the file number ten and the top 3 files
c = 0
Y_set = []
cosine_similarity = []
text_file = open('C:/Users/akhil/Downloads/Assignment1_data/Assignment1_data/'+file_name[9], 'r')
words = data_processing(text_file)
for word in words:
    Y_set.append(word)
Y_set = set(Y_set)

def vectorized_file(X_set,Y_set,rvector): 
    list1 =[]
    list2 =[]
    for w in rvector:
        if w in X_set: 
            list1.append(1) 
        else: 
            list1.append(0)
        if w in Y_set: 
            list2.append(1)
        else: 
            list2.append(0)
    return list1,list2
    
for i in range(len(file_name)):
    word_list = []
    text_file = open('C:/Users/akhil/Downloads/Assignment1_data/Assignment1_data/'+file_name[i], 'r')
    words = data_processing(text_file)
    for word in words:
        word_list.append(word)
    word_list = set(word_list)
    rvector = word_list.union(Y_set)#combining all words with the words from the tenth file
    list1,list2 = vectorized_file(word_list,Y_set,rvector)#vectorzing and dividing into two lists to find the relation between similar and dissimilar files
    for i in range(len(rvector)):
        c+= list1[i]*list2[i]
    cosine_value = c / float((sum(list1)*sum(list2))**0.5)
    value =cosine_value
    cosine_similarity.append(value)
```
